```python
# markers
# keyword "marker" can be used within plot() function to specify the marker type
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y, marker = "o")
# if the marker keyword is not specified, it will be considered as points plotting style and connecting lines won't get plotted
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:r') 
# here, 'o:r' represent the marker|line|color combincation - this is called format strings
plt.show()
```

```python
# line values are "-" for solid line, ":" for dotted line, "--" for dashed line, "-." for dashed/dotted line
# if the line value in the fmt parameter is left empty, no line will be plotted
# color values are - "r" for red, "g" for green, "b" for blue, "c" for cyan, "m" fo magenta, "y" for yellow, "b" for black, and "w" for white
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o-.m') 
plt.show()
```

```python
# a keyword argument markersize or it's short form "ms" can be used to determind markersize
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:r', ms = 20) 
plt.show()
```

```python
# a keyword argument markeredgecolor or mec can be used to specify the border color for the marker
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:m', ms = 20, mec = "r") 
plt.show()
```

```python
# markerfacecolor or mfc can be used for face of the marker
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:r', ms = 20, mec = "b", mfc = "c") 
plt.show()
```

```python
# fontdict() parameter can be used to define fonts for each of these labels and title
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# labels parameter can be used to label the pie chart. Labels should be an array
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
# the default start angle is at the x-axis, but this can be changed by specifying a startangle parameter
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
# to make some wedges stand out, explode parameter can be used
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
# by setting shadow parameter to true, shadow can be shown
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
# similar to other charts that have been discussed, custom colours can be used
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
# legend() function can be used to show legend
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
# title parameter helps to add a header
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
